```python
import requests
import pandas as pd

# Consumir API de Rick and Morty (primeros 50 personajes)
url = "https://rickandmortyapi.com/api/character?page=1"
response = requests.get(url)
data = response.json()

# Extraer resultados
personajes = data['results']

# Normalizar a DataFrame
df_rm = pd.json_normalize(personajes)
df_rm.head()
```

```python
# Quitar columnas con valores nulos
df_rm_limpio = df_rm.dropna(axis=1)

# También podemos dejar solo las columnas que nos interesan:
df_rm_limpio = df_rm_limpio[[
    'id', 'name', 'status', 'species', 'gender', 'origin.name', 'location.name'
]]

df_rm_limpio.head()
```

```python
df_rm_limpio = df_rm_limpio.copy()
df_rm_limpio['universo'] = 'Rick and Morty'
df_rm_limpio.head()
```

```python
import os

# Obtener ruta del escritorio automáticamente
escritorio = os.path.join(os.path.expanduser("~"), "Desktop")

# Crear el archivo CSV en esa ruta
ruta_csv = os.path.join(escritorio, "rick_and_morty_personajes.csv")
df_rm_limpio.to_csv(ruta_csv, index=False)

print(f"Archivo guardado en: {ruta_csv}")
```
