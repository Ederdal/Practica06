### 1. Importacion de Datos de una Base de datos de MySQL

```python
import pandas as pd
from sqlalchemy import create_engine

# Parámetros de conexión
usuario = 'root'
contraseña = 'edgar2411'
host = 'localhost'
puerto = '3309'
basedatos = 'hospital'
tabla = 'tbb_personas'

# Crear motor de conexión
conexion = create_engine(f'mysql+pymysql://{usuario}:{contraseña}@{host}:{puerto}/{basedatos}')
```

```python
pip install pymysql
```

```python
# Leer los datos usando la conexión
df = pd.read_sql(f"SELECT * FROM {tabla}", con=conexion)
df.head()
```

### 2. Limpiar Datos (Cualquier metodo de limpieza drop)

```python
# Eliminar columnas que tengan al menos un valor nulo
df_limpio = df.dropna(axis=1)
df_limpio.head()
```

```python
# Hacer una copia segura antes de modificar
df_limpio = df.dropna(axis=1).copy()
```

### 3. Transfromación de datos. 

```python
df_limpio['categoria'] = 'Ejemplo'
df_limpio.head()
```

### 4. Exportar Datos a un Tabla temporal de SQL (MySQL)

```python
df_limpio.to_sql('tabla_temporal', con=conexion, if_exists='replace', index=False)
print("Exportación completa")
```
